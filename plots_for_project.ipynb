```python
import numpy as np
from scipy.stats import norm as normal
import matplotlib.pyplot as plt
from scipy.stats import gamma
```

```python
fig, ax = plt.subplots(1, 1)
a = 4
x = np.linspace(gamma.ppf(0.01, a),
                gamma.ppf(0.99, a), 100)

ax.plot(x, gamma.pdf(x, a))
plt.show()
```

```python
#t = lambda c, T_b: c/T_b
t = lambda c, T_s: c*np.exp(-T_s/(20*2))
c = 185*5.5
x = np.linspace(170,250)
plt.plot(x, t(c,x), label = r"$1017.5\cdot exp(\frac{-T_{s}}{40})$")
plt.title(r"Seconds to Burn After Popping")
plt.xlabel(r"$Temperature\ (Celsius)$")
plt.ylabel(r"$Time\ (Seconds)$")
plt.legend(prop={'size': 12})
plt.grid()
plt.show()
```
